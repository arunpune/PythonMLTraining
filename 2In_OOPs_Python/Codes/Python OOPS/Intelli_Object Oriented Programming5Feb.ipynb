# Overloading a Function

#Overloading is passing multiple  or different number of parameters to a function 

```python
#we cannot directly use overloaind in Python
def product (a,b):
    p = a*b
    print(p)
    
def product (a,b,c):
    p = a*b*c
    print(p)
    
```

```python
#THe following code will not work because we need a work around to implement polymorphism
#product(2,3)

product(4,5,2)
```

We need to use a workaround to implement polymorhism in python

*args is used implement overloading in teh python
if the parameters are passed as an interger they should add
if they are passed as string the values should concat

```python
def productpoly(datatype, *args):
    if datatype =='int':
        answer =0
    if datatype =='str':
        answer =''
        
    for x in args:
        answer = answer + x
        
    print(answer)
    
    
```

```python
productpoly('int', 5,6)
```

```python
productpoly('str','hi ', 'class')
```

# Overiding a function

Overiding is used when you want to have your own implementation in the child classes. 

```python
class A:
    def sayhi(self):
        print('I am in A')
#class A is inheriting Class B     
class B(A):
    def sayhi (self):

        print('I am in B')
        
```

```python
objB = B()
objB.sayhi()
```

# Real life Example for using Overloading

Customers Fixed Deposit Example

```python
class Customer:
    def CustFD(self):
        print('I am Fixed Deposit Formula For Customer')
    def CustLongtermFD(self):
        print('I am Cust LongTerm Fixed Deposit Formula For Customer')
        
    #100 lines of code
#class A is inheriting Class B     
#class A is inheriting Class B     

```

```python
objCust = Customer()
objCust.CustFD()
```

```python
class PremiumCustomer(Customer):
    def CustFD(self):
        print('I am Premium Fixed Deposit Formula For PremiumCustomer')
    def CustPremiumLongtermFD(self):
        print('I am CustPremiumLongtermFD Formula For PremiumCustomer')
```

```python
objPreCust = PremiumCustomer()
objPreCust.CustLongtermFD()
objPreCust.CustFD()
```
