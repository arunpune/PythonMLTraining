# How to create a Series Object from a List of Integers

```python
import numpy as np
import pandas as pd
```

```python
data =[1,2,3,4]

print(data)
print (type(data))
print ("***** numpy object *****")
numpy1 = np.array(data)
print(numpy1)
print (type(numpy1))
print ("***** pandas series object *****")

#Series object
series1 = pd.Series(data)
print(series1)
print (type(series1))
```

# Concat

# Concat the data from 2 different data source

here we will create a dataframe from key value pairs


```python
dict3 = {'A':['A0','A1'],
         'B':['B0','B1'],
         'C':['C0','C1'],
         'D':['D0','D1']
        }


dict4 = {'A':['A0','A1'],
         'B':['B0','B1'],
         'C':['C0','C1'],
         'D':['D0','D1']
        }


df7 = pd.DataFrame(dict3)
print(df7)
print('******')
df8 = pd.DataFrame(dict4)
print(df8)
```

```python
pd.concat([df7,df8])
```

# How to CONCAT Dataframes with no common column names

```python
dict5 = {'A':['A0','A1'],
         'B':['B0','B1'],
         'C':['C0','C1'],
         'D':['D0','D1']
        }


dict6 = {'AA':['A0','A1'],
         'BB':['B0','B1'],
         'CC':['C0','C1'],
         'DD':['D0','D1']
        }



df9 = pd.DataFrame(dict5)
print(df9)
print('******')
df10 = pd.DataFrame(dict6)
print(df10)
print('******')
pd.concat([df9,df10])
```

Let us change the Index values of the data

```python
dict7 = {'A':['A0','A1'],
         'B':['B0','B1'],
         'C':['C0','C1'],
         'D':['D0','D1']
        }


dict8 = {'AA':['A0','A1'],
         'BB':['B0','B1'],
         'CC':['C0','C1'],
         'DD':['D0','D1']
        }



df11 = pd.DataFrame(dict7,index=[1,2])
print(df11)
print('******')
df12 = pd.DataFrame(dict8,index=[3,4])
print(df12)
print('******')
pd.concat([df11,df12])
```

# Merge the data using SQL like script

Merge works like an INNER JOIN

```python
dict9 = {'Key':['K1','K2'],
         'B':['B1','B2'],
         'C':['C1','C2'],
         'D':['D1','D2']
        }


dict10 = {'Key':['K1','K3','K4'],
         'BB':['B1','B3','B4'],
         'CC':['C1','C3','C4'],
         'DD':['D1','D3','D4']
        }


df13 = pd.DataFrame(dict9)
df14 = pd.DataFrame(dict10)
df15 = pd.merge(df13,df14)
print(df15)
```

```python
dict9 = {'Key':['K1','K2'],
         'B':['B1','B2'],
         'C':['C1','C2'],
         'D':['D1','D2']
        }


dict10 = {'Key':['K1','K3','K4'],
         'BB':['B1','B3','B4'],
         'CC':['C1','C3','C4'],
         'DD':['D1','D3','D4']
        }


df13 = pd.DataFrame(dict9)
df14 = pd.DataFrame(dict10)

print (df13)
print (df14)
print('*****inner*****default')
df15 = pd.merge(df13,df14)
print(df15)

print('*****inner*****')
df15inner = pd.merge(df13,df14, how='inner', on ='Key')
print(df15inner)


```

```python
print (df13)
print (df14)

print('*****outer*****')

df16 = print(pd.merge(df13,df14,how='outer', on ='Key'))
```

```python
print('*****left*****')
df17 = pd.merge(df13,df14,how='left', on ='Key')
print(df17)
```

# What values will we get if we do RIGHT JOIN

```python
print('*****right*****')
df18 = pd.merge(df13,df14,how='right', on ='Key')
print(df18)
```

```python
print('*****cross*****')
df19 = pd.merge(df13,df14,how='cross', on ='Key').drop('Key')
```

```python
print('*****cross*****')
df20 = pd.merge(df13,df14,how='outer', on ='Key')
print(df20)
```
