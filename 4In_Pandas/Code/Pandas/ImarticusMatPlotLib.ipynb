# legend

```python
x=[1,2,3,4]
y=[5,6,7,8]

plt.plot(x,y, color='green',linewidth=2,linestyle='-')
my_labels=['line']
plt.legend(labels=my_labels,fontsize=10)
```

```python
#import numpy as np
x=[1,2,3,4]
y=[5,6,7,8]
plt.plot(x,y, color='green',linewidth=2,linestyle='-')
my_labels=['line']
plt.legend(labels=my_labels,fontsize=10)
plt.title(("This is income vs age"))
plt.ylabel('income',fontsize=10)
plt.xlabel('age',fontsize=10)
plt.grid(True)
plt.xticks([0,1,2,3],["a","b","c","d"])
plt.yticks([0,2,4,6],["a","b","c","d"])
#plt.xticks(np.arange(4),["a","b","c","d"])
#plt.yticks(np.arange(4),["a","b","c","d"])
plt.show()
plt.savefig('imarplot.jpg')
```
