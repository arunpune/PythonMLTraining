```python
import pandas as pd
```

```python
'''
1. Write a function that takes start and end of a range returns a Pandas series object containing numbers within that range.
   In case the user does not pass start or end or both they sould default to 1 and 10 respectively. EG.:
   range_series() -> Should Return a pandas series from 1 to 10
   range_series(5) -> Should Return a pandas series from 5 to 10
   range_series(5, 10) -> Should Return a pandas series from 5 to 15
'''
def range_series(start = 1, end = 10):
    return pd.Series(list(range(start, end + 1)))
```

```python
print(range_series())
print(range_series(5))
print(range_series(5, 15))
```

```python
'''
2. Create a function that takes in two lists named keys and values as arguments.
   Keys would be strings and contain n string values.
   Values would be a list containing n lists.
   The methods should return a new pandas dataframe with keys as column names and values as their corresponding values
   e.g. ->
   create_dataframe(["One", "Two"], [["X", "Y"], ["A", "B"]]) -> should return a dataframe
   
      One Two
    
    0  X   A
    
    1  Y   B
   
'''

def create_dataframe(keys, values):
    data_dict = { keys[index]: values[index] for index in range(0, len(keys)) }
    return pd.DataFrame(data_dict)
```

```python
create_dataframe(["One", "Two"], [["X", "Y"], ["A", "B"]])
```

```python
'''
3. Create a function that concatenates two dataframes.
   Use previously created function to create two dataframes and pass them as parameters
   Make sure that the indexes are reset before 
'''
def concat_data_frames(df1, df2):
    return pd.concat([df1, df2]).reset_index(drop=True)

concat_data_frames(
    create_dataframe(["One", "Two"], [["X", "Y"], ["A", "B"]]),
    create_dataframe(["One", "Two"], [["V", "W"], ["C", "D"]])
)
```

```python
'''
5. Write a method that will take a column name as argument and return the name of the column with which the
   given column has the highest correlation.
   The data to be used is the cars dataset.
   The returned value should not the column named that was passed as the parameters.
   E.G: get_max_correlated_column('mpg') -> should return 'drat'
'''
def get_max_correlated_column(col_name):
    cars_df = pd.read_csv('cars.csv')
    correlation_matrix = cars_df.corr()
    return correlation_matrix[col_name].sort_values(ascending=False).index[1]
```

```python
get_max_correlated_column('mpg')
```
