```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    
    def __str__(self):
        return "Name: {0}, Salary: {1}".format(self.name, self.salary)
```

```python
emp = Employee("Employee #1", 100.00)
```

```python
print(emp)
```

```python
class Calculator:
    def _add(self, x, y):
        return x + y
    
    def _sub(self, x, y):
        return x - y
    
    def _mul(self, x, y):
        return x * y
    
    def _div(self, x, y):
        return x / y
    
    def execute_command(self, command, x, y):
        if command.lower() == "add": return self._add(x, y)
        elif command.lower() == "sub": return self._sub(x, y)
        elif command.lower() == "mul": return self._mul(x, y)
        elif command.lower() == "div": return self._div(x, y)
```

```python
obj = Calculator()
```

```python
print(obj.execute_command("Add", 2, 3))
print(obj.execute_command("sub", 2, 3))
print(obj.execute_command("mUl", 2, 3))
print(obj.execute_command("div", 2, 3))
```
