```python
'''
    Given an html file containing blog posts
        - Parse the file using BeautifulSoup
        - Extract the Title, Summary and Link of all the posts
        - Save all this data in a csv file named 'scraped_data.csv'
'''
```

```python
from bs4 import BeautifulSoup
import csv
```

```python
with open('index.html') as content:
    soup = BeautifulSoup(content, 'lxml')
```

```python
all_articles = soup.find_all('div', class_='post')
```

```python
csv_file = open('scraped_data.csv', 'w+', newline='')
writer = csv.writer(csv_file)
```

```python
writer.writerow(['Headline', 'Link', 'Summary'])
```

```python
for article in all_articles:
    headline = article.h3.a.text
    link = article.h3.a["href"]
    link = 'http://localhost:3000/{0}'.format(link)
    summary = article.p.text
    writer.writerow([headline, link, summary])
```

```python
csv_file.close()
```
