####  Read the data given in both ratings and movies files. 

```python
import pandas as pd
movies = pd.read_csv("D:/DSA-M/ML with Python/projects/MOVIE-RECOMMENDATION/movies.csv")
ratings = pd.read_csv("D:/DSA-M/ML with Python/projects/MOVIE-RECOMMENDATION/ratings.csv")
display(movies.head())
display(ratings.head())
```

```python
print(movies.shape)
print(ratings.shape)
```

#### Merge those datasets into one

```python
merged=pd.merge(movies, ratings, on='movieId')
display(merged.head(10))
```

#### Extract columns with useful information

```python
merged = merged[['movieId', 'title', 'userId', 'rating']]
display(merged.head(10))
```

#### Generate a pivot table.

```python
usertable=pd.pivot_table(merged, index=['userId'], values=['rating'], aggfunc=[len], margins=True)
usertable
```

```python
import numpy as np

movietable=pd.pivot_table(merged, index=['title'], values=['rating'], aggfunc=[len, np.mean], margins=True)
movietable
```

```python
print("UserId - Number of ratings provided")
merged['userId'].value_counts()
```

```python
print("MovieId - Number of ratings provided")
merged['movieId'].value_counts()
```

```python
print("Movies with ratings  > 4.0")
display(pd.unique(merged[merged.rating > 4.0]['title']))
```

#### Perform data cleaning by dropping users with less than 10 reviews. 

```python
print(merged.groupby(['userId']).agg({'rating':len}).reset_index().sort_values('rating'))
print("No users have less than 10 reviews")
```
