```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
'''
    1. Load cars data as dataframe using pandas and create a bar plot between number of 
    cylinders and frequncy of cars with that 
    many number of cylinders.
       
    Draw a bar plot.
'''


```

```python
data = pd.read_csv('cars.csv')
```

```python

#data['cyl'].plot.bar()
data['cyl'].value_counts()
data['cyl'].value_counts().plot(kind='bar')
```

```python
'''
    2.	Write code to load data from cars and print a bar graph of count of columns with null values.
The bar graph should the column names on the x-axis and the frequency of the null values on the y-axis

'''
ans = data.isnull().sum()
#ans.index
#ans.values
#print(ans)

plt.bar(ans.index, ans.values)
```

```python
'''
    3. Use the 'mpg' (Miles Per Gallon column) and draw a histogram
      i. Set xlabel: Miles per gallon
      ii. Set ylabel: Frequency
      iii. Set title as Miles Per Gallon Histogram
      iv. Use mpg column to generate a histogram
'''
plt.xlabel('Miles per gallon')
plt.ylabel('Frequency')
plt.title('Miles Per Gallon Histogram')
plt.hist(data.mpg)
```

```python
'''
    4. Draw a boxplot for the  hp column
      i. Set xlabel: Car Horsepower
      ii. Set the tick label: hp
      iii. Set title as :Boxplot for car horsepower
      iv. Use hp column to generate a boxplot
      v. box plot should show the mean triangle icon
      
'''
plt.title('Boxplot for car horsepower')
plt.xlabel('Car Horsepower')
plt.boxplot(data['hp'], showmeans=True)

#plt.boxplot(data['hp'], showmeans=True)

#data.loc[:,"mpg":"cyl"]
plt.xticks([1],['hp'])
plt.show()
```

```python
#5
'''draw a scatter plot with 3 variables: mpg, hp, disp.
(mpg) should be on the X-Axis.
(hp) should represented as a data point with a marker
(disp) should represented as a data point with a different marker
x label should be: mpg
y label should: Y-axis
show the grid
show the legend for : hp, disp
show the Title : "mpg-hp-disp scatter plot"    
'''

####################
plt.scatter(data.mpg,data.hp, c='g', s=200,edgecolors='r',marker='o',alpha =0.5,label='hp')
plt.scatter(data.mpg,data.disp,c='r', s=400,edgecolors='b',marker='1',alpha =1,label='disp')
plt.legend()
####################

plt.title("mpg-hp-disp scatter plot")
plt.xlabel("mpg")
plt.ylabel("Y-Axis")
plt.grid(True)
#plt.savefig('ScatterPlot2.png')
plt.show()
```
