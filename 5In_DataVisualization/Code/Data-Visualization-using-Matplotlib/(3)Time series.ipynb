#### Importing Libraries and Data

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
#stock_data = pd.read_csv('datasets/stocks.csv')
stock_data = pd.read_csv('datasets/stocks - Time.csv')
```

```python
stock_data.head()
type(stock_data.Date)
```

```python
#stock_data['Date'] = pd.to_datetime(stock_data['Date'])
stock_data['DateTime'] = pd.to_datetime(stock_data['DateTime'])
stock_data.head()

```

#### Compare Stock side-by-side

```python
plt.plot(stock_data['DateTime'],
         stock_data['AAPL'],
         color='green')
#plt.xticks('DateTime')
plt.xticks(stock_data['DateTime'])
plt.show()
```

```python
fig = plt.figure(figsize=(10,6))


    
    
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.05, 0.65, 0.5, 0.3])

ax1.set_title('AAPL vs IBM(inset)')

ax1.plot(stock_data['Date'],
         stock_data['AAPL'],
         color='green')

ax2.plot(stock_data['Date'],
         stock_data['IBM'], 
         color='blue')

plt.show()
```
