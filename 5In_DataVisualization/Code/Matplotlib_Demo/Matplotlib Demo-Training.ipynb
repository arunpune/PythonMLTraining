# Total Order Discount

```python
import matplotlib.pyplot as plt
%matplotlib inline

#Data Preparation

total = [20,4,1,30,20,10,20,70,30,10]
order = [10,3,1,15,17,2,30,44,2,1]
discount = [30,10,20,5,10,20,50,60,20,45]
data = list([total, order,discount])
print(data)
#Plotting the data
plt.boxplot(data, showmeans = True)

plt.title("Box Plot Demo")
plt.xticks([1,2,3],['total','order','discount'])
plt.grid(True)
plt.show()
```

# How to create a violin plot?

```python
import matplotlib.pyplot as plt
%matplotlib inline

#Data Preparation
total = [20,4,1,30,20,10,20,70,30,10]
order = [10,3,1,15,17,2,30,44,2,1]
discount = [30,10,20,5,10,20,50,60,20,45]
data = list([total, order,discount])
#Plotting the data
plt.violinplot(data, showmeans = True, showmedians = True)
plt.title("Violin Plot Demo")
plt.xlabel("X")
plt.ylabel("Y")

plt.xticks([1,2,3],['total','order','discount'])

plt.grid(True)
plt.grid(True)
plt.show()
```

# How to create a Pie Chart

```python
#importing the required libraries
import matplotlib.pyplot as plt
%matplotlib inline

#preparing the dataset
label = ['Dog','Cat','Wolf','Lion']
sizes = [50,45,60,80]

#creating the pie chart 
#plt.pie(sizes, labels = label,autopct='%1.1f%%',shadow=True)
plt.pie(sizes, labels = label,autopct='%1.1f%%',shadow=True, startangle = 90,explode = (0,0.5,0.5,0))
#plt.pie(sizes, labels = label,autopct='%1.1f%%',shadow=True, startangle = 90)


plt.title("Pie Chart Demo")

plt.show()
```

# How to create a Area Chart or Stack Chart or Stack Plot?
