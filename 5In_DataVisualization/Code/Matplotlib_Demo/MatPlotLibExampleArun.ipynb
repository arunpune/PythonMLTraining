```python
plt.figure(figsize=(5,5))
plt.scatter (1,10)
```

```python
for i in range (1,5):
    plt.subplot (2,2,i)
    plt.text(0.5,0.5,str((2,2,i)),ha='center', fontsize =10)
    plt.tight_layout()
```

```python
x=[0,10,15,13]
y=[0,11,14,15,]
plt.plot(x,y,color='blue',linewidth=2,linestyle="-.")
plt.plot(x,y,alpha =.1)
plt.ylabel('Y')
plt.xlabel('X')
plt.legend('X');
```
