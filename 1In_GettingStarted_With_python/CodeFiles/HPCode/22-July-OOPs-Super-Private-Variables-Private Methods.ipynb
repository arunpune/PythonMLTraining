# Super()

How to use Super() to call the parent class in our sub class?

Why should we use Super() to call the parent class in our sub class?

```python
#class Rectangle:

#class Square:
```

```python
class Rectangle:
    def __init__( self,length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * self.length + 2 * self.width
```

```python
class Square:
    def __init__(self, length):
        self.length = length

    def area(self):
        return self.length * self.length

    def perimeter(self):
        return 4 * self.length
```

```python
#rectange =  Rectangle
square = Square(4)
square.area()
```

```python

rectangle = Rectangle(2,4)
rectangle.area()
```

# Use Inheritance

Single inheritance, <br> user super key word, <br> init constructor

```python
class SuperRectangle2:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * self.length + 2 * self.width

```

```python
# Here we declare that the Square class inherits from the Rectangle class
class Square2(SuperRectangle2):
    def __init__(self, length):
        super().__init__(length, length)
```

```python
square2 = Square2(4)
square2.area()
```
