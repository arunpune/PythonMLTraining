# List comprehension for loop

```python
#  list comprehension in Python 

# Create a  list  that contains square of all odd numbers from 
# range 1 to 10 using list comprehension 
odd_square = [x ** 2 for x in range(1, 11) if x % 2 == 1] 
print (odd_square) 

#x ** 2 - output expression
#x- input variable 
#range(1, 11) - input data 
# if x % 2 == 1 - test expression
```

for understanding, above generation is same as when we implement the same condition using regular form loops

# normal for loop

```python
# Create a  list  that contains square of all odd numbers from 
# range 1 to 10 using a normal for loop 

odd_square = [] 
for x in range(1, 11): 
	if x % 2 == 1: 
		odd_square.append(x**2) 
print (odd_square )

```

```python
# list for lowering the characters 
print( [x.lower() for x in ["A","B","C"]] )

#x.lower() -output expression
#x input variable
#["A","B","C"] input range
```

```python
# list which extracts number 
string = "my phone number is : 11122 !!"
print("\nExtracted digits") 
numbers = [x for x in string if x.isdigit()] 
print (numbers )

#x - output expression
#x - input variable
#string - input
#x.isdigit() - test
```

# create a dictionary using list comprehension 

```python
# Python code to demonstrate dictionary 
# creation using list comprehension 
#Output expression is in Key:value format

myDict = {x: x**2 for x in [1,2,3,4,5]} 
print (myDict) 
```

```python
sDict = {x.upper(): x*3 for x in 'coding '} 
print (sDict) 
```

```python
# Python code to demonstrate dictionary 
# comprehension using if. 
newdict = {x: x**3 for x in range(10) if x**3 % 4 == 0} 
print(newdict) 
```

# Static Methods in Python

```python

# use of  static method. 
#in Static methods we do not need to create a object to call a method.
class Person: 
    def __init__(self, age): 
        self.age = age 
    # a static method to check if a Person is adult or not. 
    @staticmethod
    def isAdult(age): 
        return age > 18

# print the result 
print (Person.isAdult(22) )

#Uses of static method
#there some function that we need to use just for the specific purpose
#-min(), interest calculations, 

```
