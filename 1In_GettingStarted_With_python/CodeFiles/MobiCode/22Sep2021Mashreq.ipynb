# performance - When to use

```python
#time taken by mutalble list
import time

def using_List():
    t1 = time.time()#Starting/Initial Time
    L=[]
    X = range(50000)
    for item in X:
        L.append(item)
    #print(L)
    return time.time()-t1

list_time = using_List()
print(list_time)

```

```python
#time taken by mutalble list
import time

def using_Tupule():
    tt1 = time.time()#Starting/Initial Time
    T=()
    X = range(50000)
    for item in X:
        T = T + (item,)
        #single tuple needs a ,
    #print(T)
    return time.time()-tt1

Tupule_time = using_Tupule()
print(Tupule_time)
```

# Pass a 'list' as an argument - Method 2- * args for iterables

```python
# sum_integers_args_3
def my_sum(*args):
    result = 0
    for x in args:
        result += x
    return result

list1 = [1, 2, 3]
list2 = [4, 5]
list3 = [6, 7, 8, 9]

print(my_sum(*list1, *list2, *list3))
```

# function that filters vowels

```python
# function that filters vowels

def fun(variable): 
	letters = ['a', 'e', 'i', 'o', 'u'] 
	if (variable in letters): 
		return True
	else: 
		return False


# sequence 
sequence = ['g', 'e', 'e', 'j', 'k', 's', 'p', 'r'] 

# using filter function
#we are creating list of vowels and filltering out only the vowels below:
filtered = filter(fun, sequence) 

print('The filtered letters are:') 
for s in filtered: 
	print(s) 
```

# Map

Use of lambda() with map()

The map() function in Python takes in a function and a list as argument. The function is called with a lambda function and a list and a new list is returned which contains all the lambda modified items returned by that function for each item. Example:

```python
# Python code to illustrate 
# map() with lambda() 
# to get double of a list. 
li = [5, 7, 22, 97, 54, 62, 77, 23, 73, 61] 
final_list = list(map(lambda x: x*2 , li)) 
print(final_list) 
```
