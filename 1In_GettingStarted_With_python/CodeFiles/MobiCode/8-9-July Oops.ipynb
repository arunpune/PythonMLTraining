Using self - is not keyword, it is a default paramater that is passed into the init(). self can can be any word. you can use "parameter" instead of self. 
in Python , the standard word is self. In Java the keyword is this.

```python
class Student ():
#class Student (object):
    def __init__(self,name,branch,year):
        self.name= name
        self.branch= branch
        self.year= year
        
    def print_details(self):
        print("Name:",self.name)
        print("Branch:",self.branch)
        print("Year:",self.year)
        
```

# Single Inheritance

```python
class fruit:
    def __init__(self):
        print("I am Fruit super class")

        
```

```python
class citrus(fruit):
    def __init__(self):
        super().__init__()
        print("I am  citrus - subclass")
```

```python
lemon = citrus()
```

```python
fruit1 = fruit()
```

# Multiple Inheritance

# pass

pass- is a Keyword that is like a place holder and it will not do any acitivity

```python
class A:
    pass
class B:
    pass

class C(A,B):
    pass
```

```python
issubclass(C,A)
```

```python
issubclass(C,B)
```

# Multilevel Inheritance

```python
class A:
    x=1
class B(A):
    y=2

class C(B):
    z=3
    
```

```python
objC= C()
```

```python
print (objC.x,objC.y,objC.z)
```

# Heirarchical Inheritance

```python
class A:
    pass
class B(A):
    pass

class C(A):
    pass
```

```python
issubclass(B,A)
```

```python
issubclass(C,A)
```

```python
issubclass(C,B)
```

# Hybrid

```python
class A:
    x=1
class B(A):
    pass

class C(A):
    pass

class D(B,C):
    pass
```
