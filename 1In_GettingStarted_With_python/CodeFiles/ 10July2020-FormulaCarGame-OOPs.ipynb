# Class Attributes vs (Object) Instance Attributes

```python
class Car:

    # create class attributes
    car_count = 0

    # create class methods. (self.name is a object attribute)
    def start(self, name, make, model):
        print ("Engine started")
        self.name = name
        self.make = make
        self.model = model
        Car.car_count += 1
```

```python
car_a = Car()

car_a.start("Corrola", "Toyota", 2015)
print(car_a.name)
print('class attribute:',car_a.car_count)
```

```python
car_b = Car()
car_b.start("City", "Honda", 2013)
print(car_b.name)
print(car_b.car_count)
```

# Inheritance

```python
# Create Class Vehicle
class Vehicle:
    def vehicle_method(self):
        print("This is parent Vehicle class method")

# Create Class Car that inherits Vehicle
class Car(Vehicle):
    def car_method(self):
        print("This is child Car class method")
```

```python
car_a = Car()
car_a.vehicle_method() # Calling parent class method
```

```python
# Create Class Vehicle
class Vehicle:
    def vehicle_method(self):
        print("This is parent Vehicle class method")

# Create Class Car that inherits Vehicle
class Car(Vehicle):
    def car_method(self):
        print("This is child Car class method")

# Create Class Cycle that inherits Vehicle
class Cycle(Vehicle):
    def cycleMethod(self):
        print("This is child Cycle class method")
```

# Method Overriding

```python
# Create Class Vehicle
class Vehicle:
    #lvehicle = 10
    def print_details(self):
        lvehicle = 1
        print("This is parent Vehicle class method")

# Create Class Car that inherits Vehicle
class Car(Vehicle):
    def print_details(self):
        print("This is child Car class method")

# Create Class Cycle that inherits Vehicle
class Cycle(Vehicle):
    def print_details(self):
        print("This is child Cycle class method")
```

```python
car_v = Vehicle()
car_v.print_details()

print(car_v.lvehicle) 


car_c = Car()
car_c.print_details()

car_cy = Cycle()
car_cy.print_details()
```

# Encapsulation

```python
# Creates class Car
class Car:

    # Creates Car class constructor
    def __init__(self, model):
        # initialize instance variables
        #self.model = model
        self.setModel(model)
    # Create  setter
    def setModel(self, model):
        if model < 2000:
            self.__model = 2000
        elif model > 2018:
            self.__model = 2018
        
    # Create getter
    def getModel(self):
        return "The car model is " + str(self.__model)
    
```

```python
carA = Car(2088)
carA
print(carA.getModel())
```
