# Multiple Arguments, *

#When you want to send unlimited number of arguments
use *args for passing unlimited arguments. instead of args you can use any varibale, *var
the value are passed as a tuple

```python



#*args takes data as tuples or values, any number of values
def fun4(*args):
    print(type(args),args)    

fun4(10,'Python',30,40)
```

```python
#*args takes data as tuples or values, any number of values
def fun4r(*args):
    return args

print(fun4r(10,'Python',30,40))
```
