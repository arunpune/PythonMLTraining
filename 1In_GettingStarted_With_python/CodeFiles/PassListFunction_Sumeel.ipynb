# Sending Multiple Parameters 

#When you want to send unlimited number of arguments<br>
use *args for passing unlimited arguments. <br>
instead of args you can use any varibale, *var the value are passed as a tuple

```python
def fun4(*args):
    print(type(args),args)    

fun4(10,'Python',30,40)
#it will print the type and values
```

# Parameters as Dictionary

##############################<br>
UNLIMITED ARGUMENTS - Key value pairs<br>
#**args takes data as key-value pairs, any number of key-values pairs, data type is dicionary<br>

##############################


```python

def fun5(**keyaw):
    print(type(keyaw),keyaw)    

fun5(a=10,b='Python',age=50)


#################################
#Output
#<class 'dict'> {'a': 10, 'b': 'Python', 'age': 50}
#################################
```
