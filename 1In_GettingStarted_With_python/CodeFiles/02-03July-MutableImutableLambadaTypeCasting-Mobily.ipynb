# 7.a)int is Immutable when use ByVal

```python
#7.a)int is Immutable. When we use ByVal-Value int is NOT CHANGED
##(ByValue)Passing Immutable Objects like Integer- value is not changed
```

```python
def increment(n):
    print("n:",n)
    print (hex(id(n)))
    n += 1
    print("n:",n)
    print (hex(id(n)))
    
    
    #though we are increment n+1 there is no change in the data
```

```python
a = 3
print("a:",a)
#printing the memory address of 3
print (hex(id(a)))
increment(a)
print("a:",a)

#memory address of a
print (hex(id(a)))
#Output
#a = 3   # a is still referring to the same object
```

# 7.b)Int is Immutable- But when used as a ByRef-Value(work around) is CHANGED by using the return value

```python
#7.b)Int is Immutable- But when used as a ByRef-Value is changed by using the return value
#Passing Immutable Objects- value is changed by using the return value

#Does that mean we will never be able to manipulate immutable objects by passing it to functions? 
#Turns out, we can still “modify” immutable objects by capturing the return of the function.
```

```python
def increment2(n):
    print("n:",n)  
    print (hex(id(n)))
  
    n += 1
    print("n:",n)
    print (hex(id(n)))
    return n

```

```python
a = 3
#memory address of a

print("a:",a)
print (hex(id(a)))
a = increment2(a)  # the return value of increment() is captured!
print("a:",a)
print (hex(id(a)))
#Output#a = 4   # a now refers to the new object created by the function. 
#new objecte created by 
```

# 7.c) List is Mutable- when used as ByRef-Value is changed. 

#7.c) List is Mutable- when used as ByRef-Value is changed. we passing a mutable object: The value is changed by using append-
#Passing Mutable Objects

```python
def increment(n):
    print('n:',n)
    print (hex(id(n)))
   # n.append([4])
    n.append(4)
    print('n:',n)
    print (hex(id(n)))
```

```python
L = [1, 2, 3]
print('L:',L)
print (hex(id(L)))
increment(L)
print('L:',L)
print (hex(id(L)))

#print (hex(id(L)))
#Output#L = [1, 2, 3, 4]   # a changed!
```

# lambada function

We use lambda functions when we require a nameless function for a short period of time. In Python, we generally use it as an argument to a higher-order function (a function that takes in other functions as arguments). Lambda functions are used along with built-in functions like filter() , map() etc

#Example 1lambada function
#################################
