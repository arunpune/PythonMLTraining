# Implementing Class with Data Encapsulation

Data Encapsulation using __(Private variables), and implementing getter and setters methods

# Getters and Setters in Python

We are not able to access the private variables directly in Python. That's why we implement the getter and setter methods.

```python
class DeSampleClass:

    def __init__(self, a):
        ## private varibale or property in Python
        self.__a = a

    ## getter method to get the properties using an object
    def get_a1(self):
        print('get is invoked')
        return self.__a

    ## setter method to change the value 'a' using an object. or set the value of a
    def set_a1(self, a):
        print('Set is invoked')
        self.__a = a
```

let us use the get and set

```python
## creating an object
objDeS = DeSampleClass(10)

## getting the value of 'a' using get_a() method
print(objDeS.get_a1())

## setting a new value to the 'a' using set_a() method
objDeS.set_a1(45)

print(objDeS.get_a1())
```

# 2 Implementing Setters with conditions in Python

Let's say if the value we passed is even and positive then we can set it to the attribute, otherwise set the value to 2.

Let's implement this by changing the set_a() method in the DESampleClass2.

```python
class DESampleClass2:

    def __init__(self, a):
        ## calling the set_a() method to set the value 'a' by checking certain conditions
        self.set_a(a)

    ## getter method to get the properties using an object
    def get_a(self):
        return self.__a

    ## setter method to change the value 'a' using an object
    def set_a(self, a):

        ## condition to check whether 'a' is suitable or not
        if a > 0 and a % 2 == 0:
            self.__a = a
        else:
            self.__a = 2
```

```python
## creating an object for the class 'SampleClass1'
obj = DESampleClass2(15)

print(obj.get_a())
```
