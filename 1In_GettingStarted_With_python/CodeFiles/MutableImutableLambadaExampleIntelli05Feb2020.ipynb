# 7.a)int is Immutable when use ByVal

```python
#7.a)int is Immutable. When we use ByVal-Value int is NOT CHANGED
##(ByValue)Passing Immutable Objects like Integer- value is not changed
```

```python
def increment(n):
    n += 1
    print (hex(id(n)))
    print("n:",n)
    
    #though we are increment n+1 there is no change in the data
```

```python
a = 3
#printing the memory address of 3
print (hex(id(a)))
increment(a)
print("a:",a)

#memory address of a
print (hex(id(a)))
#Output
#a = 3   # a is still referring to the same object
```

# 7.b)Int is Immutable- But there is work around to use it as ByRef-Valueis CHANGED by using the return value

```python
#7.b)Int is Immutable- But when used as a ByRef-Value is changed by using the return value. The value is updated by Memory Address is changed
#Passing Immutable Objects- value is changed by using the return value

#Does that mean we will never be able to manipulate immutable objects by passing it to functions? 
#Turns out, we can still “modify” immutable objects by capturing the return of the function.
```

By assigning the return value of the function to "a", we have reassigned a to refer to the new object with the value 4 created in the function. 
Note the object a initially referred to never change — it is still 3 — but by having a point to a new object created by the function, we are able to “modify” the value of a.


def increment(n):
    n += 1
    return n



```python
a = 3
#memory address of a

print("a:",a)
print (hex(id(a)))
a = increment(a)  # the return value of increment() is captured!
print("a:",a)
print (hex(id(a)))
#Output#a = 4   # a now refers to the new object created by the function
print(a)
```

# 7.c) List is Mutable- when used as ByRef-Value is changed. 

```python
#7.c) List is Mutable- when used as ByRef-Value is changed. we passing a mutable object: The value is changed by using append-
#Passing Mutable Objects
```

```python
def increment(n):
   #n.append([4])
   n.append(4)
```

```python
L = [1, 2, 3]
print('L:',L)
increment(L)
print (hex(id(L)))
print('L:',L)
print (hex(id(L)))
#Output#L = [1, 2, 3, 4]   # a changed!
```
