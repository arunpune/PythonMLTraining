# 1. User-defined function to add_10()

1.	Create a user-defined function “add_10()“ which will add 10 more to whatever value you pass in the function.

```python
# Defining an user-defined function that increments number by 10
def add_10(x):
# Using try-except to check whether a number is passed
# If value other than number is passed, error message is shown
    try:
        a=float(x)
        a = a + 10
        return a
    except:
        print("ERROR!...Enter a number")
    
a=input("Enter a value:")
result = add_10(a)
print(result)
```

# 2. User-defined function to check even/odd

2.	Create a user-defined function “check_odd_even()” which will check whether the given integer value is odd or even. If the number is odd, it will print “Odd number”, else it will print “Even number”

```python
# Declaring a function that checks for even or odd value
def check_even_odd(x):
    if(x % 2 == 0):
        print(f"The number {x} is even")
    else:
        print(f"The number {x} is odd")

# Enter integer number to find whether even or odd        
try:
    a = int(input("Enter a number:"))
    result = check_even_odd(a)
except:
    print("Error!...enter a integer number")
```
