# 1. Comparing corresponding elements of List1 and List2

```python
# Declaring two lists
L1 = [1,2,3,4,5,6]
L2 = [6,5,4,3,2,1]

# Using for to compare the lists element-by-element 
# and printing whether ith value of list1 is greater then or less than ith value of List2
for i in range(len(L1)):
    if(L1[i]>L2[i]):
        print(f"L1[{i}]:{L1[i]} greater than L2[{i}]:{L2[i]}")
    else:
        print(f"L1[{i}]:{L1[i]} less than L2[{i}]:{L2[i]}")
```

# 2. Adding 5 to each elements of the tuple using loop

```python
# Declaring tuple
tup = (10,20,30,40,50)

#Printing the elements of tuple
print("Before addition",tup)

#Converting the tuple to list 
lst=list(tup)
#print(lst)

# Incrementing each element of list by 5 using for loop
for i in range(len(lst)):
    lst[i]=lst[i] + 5
    
# Again converting the list to tuple
tup = tuple(lst)  

#Printing the output tuple
print("After addition",tup)
```

# 3. User-defined function to add_10()

```python
# Defining an user-defined function that increments number by 10
def add_10(x):
# Using try-except to check whether a number is passed
# If value other than number is passed, error message is shown
    try:
        a=float(x)
        a = a + 10
        return a
    except:
        print("ERROR!...Enter a number")
    
a=input("Enter a value:")
result = add_10(a)
print(result)
```

# 4. User-defined function to check even/odd

```python
# Declaring a function that checks for even or odd value
def check_even_odd(x):
    if(x % 2 == 0):
        print(f"The number {x} is even")
    else:
        print(f"The number {x} is odd")

# Enter integer number to find whether even or odd        
try:
    a = int(input("Enter a number:"))
    result = check_even_odd(a)
except:
    print("Error!...enter a integer number")
```
