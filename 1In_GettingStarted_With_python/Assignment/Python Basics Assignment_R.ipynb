1.) Create two lists L1 & L2, where L1 has values ->(1,2,3,4,5,6) and L2 has values ->(6,5,4,3,2,1).
Compare the corresponding element values of both the lists and print "L1 element values is greater than L2 element value", if corresponding value from L1 is greater else print "L1 element values is less than L2 element value", if corresponding value from L1 is less than L2

```python
l1 = [1,2,3,4,5,6]
l2 = [6,5,4,3,2,1]

for i in range(len(l1)):
    if l1[i]>l2[i]:
        print("L1 element values is greater than L2 element value")
    elif l1[i]<l2[i]:
        print("L1 element values is less than L2 element value")
```

2.) Create a tuple with values->(10,20,30,40,50).
Use a loop to add 5 more to each individual element of the tuple.

3.) Create a user defined function "add_10()" which will add 10 more to whatever value you pass in the function

```python
def add_10(x):
    if isinstance(x, (int,float)):
        return x+10

print(add_10(5))
print(add_10(23.2))
print(add_10('a'))
```

4.) Create a user-defined function "check_odd_even()" which will check whether the given integer value is odd or even. If the number is odd, it will print "Odd number", else it will print "Even number" 

```python
def check_odd_even(x):
    if (isinstance(x,int) and x%2!=0) :
        print(x,'is odd number')
    elif (isinstance(x,int) and x%2==0) :
        print(x,'is even number')
    else:
        print(x,'is not an integer')

check_odd_even(32)
check_odd_even(5)
check_odd_even(32.5)
```
