# webscrapping

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
html = urlopen("http://www.hubertiming.com/results/2018Resolution")
```

```python
print(html)
```

```python
#bsobj
bsobj = BeautifulSoup(html, 'lxml')
print(bsobj)
```

```python
#bsobj.fin
```

```python
table_thead= bsobj.findAll('thead')
table_thead
```

```python
table_rows= bsobj.findAll('tr')
print(table_rows)
```

```python
for row in table_rows:
    each_row= row.findAll('td')
    print(each_row)
```

```python
lists_of_rows = []
for row in table_rows:
    each_row= row.findAll('td')
    str_row= str(each_row)
    row_text = BeautifulSoup(str_row, "lxml").get_text()
    lists_of_rows.append(row_text)
#print(each_row)
print(lists_of_rows)
```

```python
import pandas as pd
import numpy as np
# the goal of this article is just to crawl the data from the website and convert it into a dataframe.
```

```python
lists_of_rows[6:]
```

```python
data = pd.DataFrame(lists_of_rows[7:])
print(type(data))
print(data)
```

```python


data.to_csv('Book1.csv', encoding='utf-8', index=False)
```
