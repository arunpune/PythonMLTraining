```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataset = pd.read_csv('salaryData.csv')
dataset
```

```python
x= pd.DataFrame(dataset.iloc[:,:-1])
print(x)

```

```python
y= pd.DataFrame(dataset.iloc[:,-1])
print(y)
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(x,y, test_size=1/3,random_state=99)
```

```python
yTest
```

```python
linearRegression = LinearRegression()
```

```python
linearRegression.fit(xTrain,yTrain)
```

```python
print(linearRegression.intercept_)#c
print(linearRegression.coef_)#slope
```

```python
#y=mx+c  ---> y= 9460.96544077 x + 26499.15318611 = using seed 3
```

```python
#y=mx+c  ---> y= 9152.98777019 x + 27751.6331058 =using seed of 99
```

```python
yPrediction =linearRegression.predict(xTest)
```
