# Random State

```python
from sklearn.model_selection import train_test_split

X_data = range(10)
y_data = range(10)
print(X_data.count)
print(y_data.count)

print("random State:11")    
print("*"*30)

for i in range(3):

    X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size =0.2,random_state = 11) # zero or any other integer
    print(y_train)
    print(y_test)
    print("*"*30)

print("random State:0")    
print("*"*30)
    
for i in range(3):

    X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size =0.2,random_state = 0) # zero or any other integer
    print(y_train)
    print(y_test)
    print("*"*30)
    

print("No random State")    
print("*"*30)

for i in range(3): 
    X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size = 0.2,random_state = None)
    print(y_train)
    print(y_test)
3
```
